```python
#def FillinMissingAge(a1 , p1): knwown as DATA IMPUtATION  # have to use list here
#     Pclass = p1 
#     age = a1
def FillinMissingAge(cols):
    Pclass = cols[1] 
    age = cols[0]
    if pd.isnull(age):   # isnull true means value is missing.
        if Pclass == 1:
            return 38
        elif Pclass == 2:
            return 30
        elif Pclass == 3:
            return 25
        else:                 # for those who have mentined wrong class and some other exceptions
            return 30
    else:
        return age    # if not missing return the same age.
```

```python
# now i want this function to apply on each line of dataset[['Age', 'Pclass']] => use of apply function[default works on axis =0 => rows]
age = dataset[['Age', 'Pclass']].apply(FillinMissingAge, axis =1 )
```
